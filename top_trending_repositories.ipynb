```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url_link = 'https://github.com/trending'

headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
web_page = requests.get(url_link, headers=headers)
print(web_page.status_code)
response = web_page.text
print(len(response))
page_info = BeautifulSoup(response, 'html.parser')
```

```python
trending = page_info.find_all('span', {'itemprop' : 'programmingLanguage'})
# top5 = trending[:5]
# print(len(trending))
# print(top5)

trending_topics = []
for topics in trending:
    trending_topics.append(topics.text)

trending_topics
```

```python
trending_names = page_info.find_all('h1', {'class' : 'h3 lh-condensed'})
trending_repo = []

for repo in trending_names:
    new_repo = repo.text.strip()
    trending_repo.append(new_repo)

repo_name = [re.sub(r'[^a-zA-Z0-9]', '', string) for string in trending_repo]

repo_name
```

```python
stars = page_info.find_all('a', {'class' : 'Link--muted d-inline-block mr-3'})
trending_stars = []

for star in stars:
    new_star = star.text.strip()
    trending_stars.append(new_star)
    main_stars = trending_stars[0::2]
    main_forks = trending_stars[1::2]

main_stars
```

```python
main_forks
```

```python
latest_stars = page_info.find_all('span', {'class' : 'd-inline-block float-sm-right'})
today_stars = []

for star in latest_stars:
    new_star = star.text.strip()
    today_stars.append(new_star)
new_stars = [re.sub(r'[^0-9]', '', string) for string in today_stars]

new_stars
```

```python
my_data = {
    'repo_name': repo_name,
    'total_stars': main_stars,
    'today_stars': new_stars,
    'total_forks': main_forks
}
```

```python
trending_repositories = pd.DataFrame(my_data)
```

```python
trending_repositories.to_csv('trending_repos.csv', index = None)
```
